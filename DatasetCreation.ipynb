```python
import os
from src.DatasetCreation import create_dataset_versions

data_dir = os.path.join(os.getcwd(), 'data')
print(f"Saving datasets to: {data_dir}")

datasets = [
    'german',
    # 'adult',
    # 'diabetes',
    # 'employment',
]

combinations = [ # Privacy distribution percentage combinations (must sum to 1.0)
    (0.66, 0.17, 0.17),
    # (0.33, 0.33, 0.34),
]

# Iterate over datasets and combinations
for dataset in datasets:
    for original_pct, generalized_pct, missing_pct in combinations:
        print(f"Creating versions for {dataset} with original={original_pct}, generalized={generalized_pct}, missing={missing_pct}")
        create_dataset_versions(
            dataset=dataset,
            original_pct=original_pct,
            generalized_pct=generalized_pct,
            missing_pct=missing_pct,
            seed=42,
            data_dir=data_dir
        )
```
